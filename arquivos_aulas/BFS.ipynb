### Grafos

```python
class Grafo:
    def __init__(self, N):
        self.N = N
        self.adj = [[] for _ in range(N)]
    
    def __repr__(self):
        return '\n'.join([str(line) for line in self.adj])
    
    def add_arrow(self, n1, n2):
        self.adj[n1].append(n2)
        
    def add_edge(self, n1, n2):
        self.add_arrow(n1, n2)
        self.add_arrow(n2, n1)
    
    def BFS(self, n):
        visitados = [False] * self.N
        fila = [n]
        
        while len(fila) != 0:
            primeiro_da_fila = fila.pop(0)
            visitados[primeiro_da_fila] = True
            
            print(f'Visitando o n√≥ {primeiro_da_fila}')
            
            for adjacente in self.adj[primeiro_da_fila]:
                if not visitados[adjacente] and not adjacente in fila:
                    fila.append(adjacente)
```

```python
grafo = Grafo(8)
```

```python
grafo.add_edge(0, 1)
grafo.add_edge(0, 3)
grafo.add_edge(1, 2)
grafo.add_edge(3, 4)
grafo.add_edge(3, 5)
grafo.add_edge(4, 5)
grafo.add_edge(4, 7)
grafo.add_edge(5, 6)
grafo.add_edge(5, 7)
grafo.add_edge(6, 7)
```

```python
grafo.BFS(0)
```
